```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import roc_curve, auc, confusion_matrix, classification_report, accuracy_score
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.feature_extraction import text
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.preprocessing import StandardScaler
# from imblearn.over_sampling import RandomOverSampler, ADASYN, SMOTE
# from imblearn.under_sampling import RandomUnderSampler
from sklearn.utils import resample
from sklearn.naive_bayes import MultinomialNB, BernoulliNB, GaussianNB


from textblob import TextBlob
import pickle
from datetime import datetime

from nltk.stem import wordnet
from nltk.stem import WordNetLemmatizer

sns.set(style = "darkgrid")
```

Join both review and metadata datasets from zip file

```python
# Consolidate review metadata for NJ, VT, CT, and PA states

reviewContent = pd.read_csv("./reviews_data/reviewContent", names = ['user_id', 'prod_id', 'date', 'review'], sep = '\t', parse_dates = ['date'])
metadata = pd.read_csv("./reviews_data/metadata",names=["user_id", "prod_id", "rating", "label", "date"], sep = '\t', parse_dates = ['date'])



reviews = pd.merge(reviewContent, metadata, how = 'left', 
                    left_on = ['user_id', 'prod_id', 'date'], right_on = ['user_id', 'prod_id', 'date'])
```

```python
# Consolidate review metadata for NY state

reviewContent_ny = pd.read_csv("./reviews_data/reviewContent_ny", names = ['user_id', 'prod_id', 'date', 'review'], sep = '\t', parse_dates = ['date'])
metadata_ny = pd.read_csv("./reviews_data/metadata_ny",names=["user_id", "prod_id", "rating", "label", "date"], sep = '\t', parse_dates = ['date'])



reviews_ny= pd.merge(reviewContent_ny, metadata_ny, how = 'left', 
                    left_on = ['user_id', 'prod_id', 'date'], right_on = ['user_id', 'prod_id', 'date'])
```

```python
reviews = pd.concat([reviews, reviews_ny]).reset_index(drop = True)
```

```python
reviews.shape
```

```python
reviews.head()
```

```python
# Convert classification of fake reviews from '-1' to '0'

reviews['label'] = reviews['label'].apply(lambda x: 0 if x == -1 else x)
```

```python
# Check label counts

reviews['label'].value_counts()
```

```python
# Percentage of fake reviews out of all reviews

1- np.mean(reviews['label'])
```

Given the significant class imbalance where ~12% of all reviews are fake (based on Yelp's proprietary algorithm), we will look to address this class imbalance to improve our predictions. 

### Due to very long processing time, we will downsample the non-fake reviews so that the overall number of reviews can be reduced to a more manageable number

```python
df_real = reviews[reviews['label'] == 1]
df_fake = reviews[reviews['label'] == 0]

print(df_real.shape)
print(df_fake.shape)
```

```python
# Downsize non-fake reviews

df_real_downsample = resample(df_real, replace = False, n_samples = 117299, random_state = 240)
reviews_df = pd.concat([df_real_downsample, df_fake])

print(reviews_df.shape)
print(reviews_df[reviews_df['label'] == 1].shape)
print(reviews_df[reviews['label'] == 0].shape)
```

```python
# Check unique user id counts for real reviews_df and fake reviews_df

print(reviews_df['user_id'][reviews_df['label'] == 1].nunique())
print(reviews_df['user_id'][reviews_df['label'] == 0].nunique())
```

## Part 2: Experiment using non-review text features to predict rating (based on feature engineering)

```python
reviews_df.head()
```

```python
# Include only non-text features
model_nontext = reviews_df[['rating','review_text_count', 'day_of_week', 'sentiment_polarity', 'subjectivity',
                            'label']]
```

```python
model_nontext.info()
```

**Testing for logistic regression**

```python
# get dummies for day of week 
model_nontext_log = pd.get_dummies(model_nontext, columns = ['day_of_week'], drop_first = True)

# Get predictor and target values and do train test split
X = model_nontext_log.drop(columns = ['label'])
y = model_nontext_log['label']

# Perform train-test-split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 22, stratify = y)
```

```python
# Standardise predictor variables

ss = StandardScaler()

Xs_train = ss.fit_transform(X_train)
Xs_test = ss.transform(X_test)
```

```python
# Initialise LogisticRegression function

logreg = LogisticRegression(solver = 'liblinear')
logreg.fit(Xs_train, y_train)

cv_logreg = cross_val_score(logreg, Xs_train, y_train, cv = 10, n_jobs = -1)
print('Cross Validation Score: {:0.3} ± {:0.3}'.format(cv_logreg.mean().round(3), cv_logreg.std().round(3)))

# On testing dataset
y_pred = logreg.predict(Xs_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
print('Confusion Matrix: ', pd.crosstab(y_test, y_pred,
                    rownames=['Actual'],
                    colnames=['Predicted'],margins=True))
print(classification_report(y_test, y_pred))
```

**Testing for Random Forest Classifier**

```python
# get dummies for day of week 
model_nontext_rfc = pd.get_dummies(model_nontext, columns = ['day_of_week'], drop_first = True)

# Get predictor and target values and do train test split
X = model_nontext_rfc.drop(columns = ['label'])
y = model_nontext_rfc['label']

# Perform train-test-split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 22, stratify = y)
```

```python
# Initialise Random Forest Classifier function

rfc = RandomForestClassifier(random_state=620)
rfc.fit(X_train, y_train)

cv_rfc = cross_val_score(rfc, X_train, y_train, cv = 10, n_jobs = -1)
print('Cross Validation Score: {:0.3} ± {:0.3}'.format(cv_rfc.mean().round(3), cv_rfc.std().round(3)))

# On testing dataset
y_pred = rfc.predict(X_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
print('Confusion Matrix: ', pd.crosstab(y_test, y_pred,
                    rownames=['Actual'],
                    colnames=['Predicted'],margins=True))
print(classification_report(y_test, y_pred))
```

Using only non-review text features, the accuracy of 0.59 for Logistic Regression is pretty poor relative to the baseline.

We will first use cross_validation on the train set to do a high level evaluation of the effectiveness of various model types and sampling methods using the function below. 

***Modelling techniques***
- 1) Random Forest Classifier
- 2) Multi Naive Bayes
- 2) Logistic Regression

**Baseline would be 50% now since we have downsampled the majority class such that there are equal samples for both majority and minority classes**

```python
# Functions for the 3 modelling techniques to be tested

# Random Forest Classifier
def RandomForest(X_train, X_test, y_train, y_test):
    t0 = datetime.now()
    
    rfc = RandomForestClassifier(n_estimators = 10, random_state = 35)
    cv_rfc = cross_val_score(rfc, X_train, y_train, cv = 3, n_jobs = -1)
    print('-'*100)
    print('RandomForestClassifier')
    print('-'*100)
    print('Cross Validation Score: {:0.3} ± {:0.3}'.format(cv_rfc.mean().round(3), cv_rfc.std().round(3)))
    print(datetime.now() - t0)
    rfc.fit(X_train, y_train)
    y_pred = rfc.predict(X_test)
    print('Accuracy Score: ', accuracy_score(y_test, y_pred))

    print('Confusion Matrix: ', pd.crosstab(y_test, y_pred,
                        rownames=['Actual'],
                        colnames=['Predicted'],margins=True))
    print('-'*100)
    print(classification_report(y_test, y_pred))
    print(datetime.now() - t0)
    print('\n')
        

# Multinomial Naive Bayes        
def MultiNaiveBayes(X_train, X_test, y_train, y_test):
    t0 = datetime.now()
    
    mnb = MultinomialNB()
    cv_mnb = cross_val_score(mnb, X_train, y_train, cv = 3, n_jobs = -1)
    print('-'*100)
    print('Multinomial Naive Bayes')
    print('-'*100)
    print('Cross Validation Score: {:0.3} ± {:0.3}'.format(cv_mnb.mean().round(3), cv_mnb.std().round(3)))
    print(datetime.now() - t0)
    mnb.fit(X_train, y_train)
    y_pred = mnb.predict(X_test)
    print('Accuracy Score: ', accuracy_score(y_test, y_pred))

    print('Confusion Matrix: ', pd.crosstab(y_test, y_pred,
                        rownames=['Actual'],
                        colnames=['Predicted'],margins=True))
    print('-'*100)
    print(classification_report(y_test, y_pred))
    print(datetime.now() - t0)
    print('\n')
    

# Logistic Regression       
def LogReg(X_train, X_test, y_train, y_test):
    t0 = datetime.now()

    logreg = LogisticRegression()
    cv_logreg = cross_val_score(logreg, X_train, y_train, cv = 3, n_jobs = -1)
    print('-'*100)
    print('Logistic Regression')
    print('-'*100)
    print('Cross Validation Score: {:0.3} ± {:0.3}'.format(cv_logreg.mean().round(3), cv_logreg.std().round(3)))
    print(datetime.now() - t0)
    logreg.fit(X_train, y_train)
    y_pred = logreg.predict(X_test)
    print('Accuracy Score: ', accuracy_score(y_test, y_pred))

    print('Confusion Matrix: ', pd.crosstab(y_test, y_pred,
                        rownames=['Actual'],
                        colnames=['Predicted'],margins=True))
    print('-'*100)
    print(classification_report(y_test, y_pred))
    print(datetime.now() - t0)
    print('\n')
```

```python
def run_pipeline(X_train, X_test, y_train, y_test):
    
    RandomForest(X_train, X_test, y_train, y_test)

    MultiNaiveBayes(X_train, X_test, y_train, y_test)

    LogReg(X_train, X_test, y_train, y_test)
```

## Test out deep learning - CNN + LSTM

```python
# Select only labels and reviews for deep learning

train = reviews_df[['label', 'review']].reset_index(drop = True)
train.head()
```

```python
import tensorflow as tf
>>> hello = tf.constant('Hello, TensorFlow!')
>>> sess = tf.Session()
>>> print(sess.run(hello))
```

```python
from keras.layers import Embedding, LSTM, Dense, Conv1D, MaxPooling1D, Dropout, Activation
from keras.models import Sequential
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
# a list contains each review as a list 
balanced_texts = []
balanced_labels = []

for i in range(len(train)):
    balanced_texts.append(train.iloc[i,1])
    balanced_labels.append(train.iloc[i,0])
```

```python
balanced_texts
```

```python
#choose the first 20000 common words and set vector size as 500
tokenizer = Tokenizer(num_words=20000)#20000
tokenizer.fit_on_texts(balanced_texts)
sequences = tokenizer.texts_to_sequences(balanced_texts)
data = pad_sequences(sequences, maxlen=500)#300
```

```python
# build model and add layers
model = Sequential()
model.add(Embedding(20000, 128, input_length=500))#20000 300
model.add(Dropout(0.2))
model.add(Conv1D(64, 5, activation='relu'))
model.add(MaxPooling1D(pool_size=4))
model.add(LSTM(128))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data, np.array(balanced_labels), test_size=0.3)
```

```python
# fit model
model.fit(data, np.array(balanced_labels), validation_split=0.4, batch_size=200, epochs=3)
#model.fit(X_train,y_train,batch_size=200)
```
